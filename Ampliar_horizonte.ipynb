## Variables

* p : horizonte de predicion
* tw : tama√±o de la ventana
* nw : numero de ventanas
* p : horizonte de predicion
* x : datos reales
* y : datos predichos
* z : datos para alimentar la red


```
import numpy as np

p = 5
nw = 4
tw = 10
x = np.array(range(30)) 


for j in range(nw):
  z = x
  for i in range(p):
    IndicePrediccion = p*j + tw + i
    IndiceInicial = (j*p) + i
    IndiceFinal = (j*p) + i + tw
    print("IndicePrediccion: ", IndicePrediccion)
    print("IndicesEntrada: ", range(IndiceInicial,IndiceFinal-1) )
    print("DatosEntrada: ", z[range(IndiceInicial,IndiceFinal)] )

    y[IndicePrediccion] = np.mean(z[range(IndiceInicial,IndiceFinal)])
    z[IndicePrediccion] = y[IndicePrediccion]

print("y: ", y)
```
